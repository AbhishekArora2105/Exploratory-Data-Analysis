```python
#data is either continuous or discrete 
# discrete data thought is numerical but treated as categorical data
# but for model building all the categorical data should be converted to numbers
```

```python
cat_features = x.select_dtypes(include = 'object').columns
num_features = x.select_dtypes(exclude = 'object').columns
```

```python
num_features
```

```python
#scaling should be done for numerical features 
#onehot encoding for categorical data
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import ColumnTransformer

#column tranformer applies tranformations to the columns of an array or dataframe
```

```python
numeric_tranformer = StandardScaler()
oh_transformer = OneHotEncoder(drop = 'first')
preprocessor = ColumnTransformer([
    ('OneHotEncoder' , oh_transformer , cat_features),('StandardScaler',numeric_tranformer,num_features)
])
```

```python
preprocessor
```

```python
x = preprocessor.fit_transform(x_train)
pd.DataFrame(x)
```
