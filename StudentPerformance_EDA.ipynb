```python
# outlier detection

#to detect >> distribution , boxplot , violinplot
```

```python
students = df.copy()
```

```python
students
```

```python
# students.drop(["G1" , "G2","G3"], axis = 1)
```

```python
def detect_outliers(columns):
    outlier_indices = []
    for col in columns:
        Q1 = students[col].quantile(0.25)
        Q3 = students[col].quantile(0.75)
        IQR = Q3 - Q1
        #mask is a boolean condition
        mask = (students[col] >= (Q1 -1.5*IQR)) & (students[col] <= (Q3+1.5*IQR))
        mask = mask.to_numpy()
        #find the  indices of values that do not satisfy the mask condition
        false = np.argwhere(~mask)
        outlier_indices.append(false)
    return np.unique(np.concatenate(outlier_indices).ravel())

#it will concatenate all the outlier index  arrays into single array and ravel flattens the multi dim array into 1 d array
```

```python
num_cols = ['age','absences']
detect_outliers(num_cols)
```
